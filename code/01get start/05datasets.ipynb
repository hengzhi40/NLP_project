```python
# 7.1 数据映射
def prefix_function(example):
    example["sentence1"] = "Prefix:" + example["sentence1"]
    return example
```

```python
all_data3 = all_data.map(prefix_function)
all_data3[:]["sentence1"]
```

```python
# 7.2 token处理的数据映射
from transformers import AutoModel, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("/gemini/pretrain2")
def token_process(example, tokenizer=tokenizer):
    model_inputs = tokenizer(example["sentence1"], example["sentence2"], max_length=128, truncation=True)
    labels = tokenizer(example["label"], max_length=128, truncation=True)
    model_inputs["labels"] = labels["input_ids"] # 对label编码之后取了inputs_ids
    return model_inputs
```

```python
all_data4 = all_data.map(token_process,  batched=True, num_proc=10) # batched, num_proc是用来处理加速的
all_data4
```
